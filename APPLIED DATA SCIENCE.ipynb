```python
import pandas as pd

# Create a DataFrame with some sample data
data = {
    "Temperature C°": [15, 20, 25, 30, 35],
    "Ice Cream Profits per hour": [100, 120, 150, 180, 200]
}
df = pd.DataFrame(data)

# Calculate the correlation coefficient between Temperature and Ice Cream Profits
correlation = df["Temperature C°"].corr(df["Ice Cream Profits per hour"])
print("The correlation between Temperature and Ice Cream Profits per hour:", correlation)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
data = {
    'Temperature C°': [39, 40, 41, 42, 43, 99, 99, 99, 101, 101],
    'Ice Cream Profits per hour': [13.17, 11.88, 18.82, 18.65, 17.02, 85.13, 87.08, 89.29, 81.91, 85.02]
}
df = pd.DataFrame(data)

# Scatter plot with regression line
plt.figure(figsize=(8, 6))  # Set the size of the plot
sns.regplot(x='Temperature C°', y='Ice Cream Profits per hour', data=df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Correlation between Temperature and Ice Cream Profits per Hour')
plt.xlabel('Temperature C°')
plt.ylabel('Ice Cream Profits per Hour')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your pandas DataFrame and 'Ice Cream Profits per hour' is the correct column
plt.figure(figsize=(8, 6))  # Set the size of the plot
sns.boxplot(y=df['Ice Cream Profits per hour'], color='lightblue')  # Correct column name
plt.title('Boxplot of Ice Cream Profits per Hour')  # Add a title
plt.show()  # Display the plot
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is stored in df
# Create the violin plot for the Ice Cream Profits per hour
plt.figure(figsize=(8, 6))  # Set the size of the plot
sns.violinplot(y=df['Ice Cream Profits per hour'], color='lightblue')  # Create the violin plot
plt.title('Violin Plot of Ice Cream Profits per Hour')  # Add a title
plt.show()  # Display the plot
```
